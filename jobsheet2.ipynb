```python
# ==============================================
# TUGAS AKHIR - MODEL D2 (Artificial Neural Network)
# ==============================================

# 1) Import library
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import Adam

# 2) Load dataset MNIST
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalisasi ke [0,1]
x_train = x_train.astype('float32') / 255.0
x_test  = x_test.astype('float32') / 255.0

# One-hot encoding label
y_train = to_categorical(y_train, 10)
y_test  = to_categorical(y_test, 10)

# 3) Bangun model D2 (ANN sederhana)
model = Sequential()
model.add(Flatten(input_shape=(28,28)))
model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(10, activation='softmax'))

# 4) Compile model
model.compile(optimizer=Adam(),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# 5) Training model
history = model.fit(x_train, y_train,
                    epochs=5,
                    batch_size=128,
                    validation_data=(x_test, y_test))

# 6) Evaluasi model
loss, acc = model.evaluate(x_test, y_test, verbose=0)
print(f"Akurasi di data test: {acc:.4f}")

# 7) Simpan model
model.save("model_d2.h5")
print("Model tersimpan sebagai model_d2.h5")

# ======================================================
# BAGIAN 2: PREDIKSI GAMBAR DARI GOOGLE DRIVE
# ======================================================
import os
from PIL import Image, ImageOps
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Path folder gambar
folder_path = "/content/drive/MyDrive/kumpulan_foto_angka"

# Fungsi preprocessing ke format MNIST
def preprocess_to_mnist_28x28(img_pil):
    img = img_pil.convert('L')            # grayscale
    img = ImageOps.autocontrast(img)
    arr = np.array(img).astype(np.uint8)

    # Invert jika background putih
    if arr.mean() > 127:
        img = ImageOps.invert(img)
        arr = np.array(img)

    # Cari bounding box digit
    thr = np.mean(arr) * 0.8
    mask = arr > thr
    if mask.any():
        ys, xs = np.where(mask)
        y0, y1 = ys.min(), ys.max()
        x0, x1 = xs.min(), xs.max()
        img = img.crop((x0, y0, x1+1, y1+1))

    # Resize ke 20x20
    img.thumbnail((20,20), Image.Resampling.LANCZOS)
    w,h = img.size

    # Pad ke 28x28
    canvas = Image.new('L', (28,28), color=0)
    canvas.paste(img, ((28-w)//2, (28-h)//2))

    arr = np.array(canvas).astype('float32') / 255.0
    arr = arr.reshape(1,28,28)
    return canvas, arr

# 8) Load kembali model D2
clf = load_model("model_d2.h5")

# 9) Prediksi semua gambar di folder (ditampilkan dalam grid 5x5)
file_list = [f for f in os.listdir(folder_path) if f.lower().endswith(('.png','.jpg','.jpeg'))]

rows, cols = 5, 5
fig, axes = plt.subplots(rows, cols, figsize=(12, 12))

results = []  # simpan hasil prediksi untuk rekap

for i, ax in enumerate(axes.flat):
    if i < len(file_list):
        fname = file_list[i]
        path = os.path.join(folder_path, fname)
        img_pil = Image.open(path)
        proc_img, arr = preprocess_to_mnist_28x28(img_pil)

        # Prediksi probabilitas
        probs = clf.predict(arr, verbose=0)[0]
        pred = np.argmax(probs)
        conf = float(np.max(probs))

        results.append((fname, pred, conf))

        ax.imshow(proc_img, cmap='gray')
        ax.set_title(f"{fname}\nPrediksi: {pred}", fontsize=10)
        ax.axis('off')
    else:
        ax.axis('off')

plt.tight_layout()
plt.show()

# ===== Rekap hasil prediksi =====
print("\nRekap Prediksi:")
for r in results:
    print(f"- {r[0]} -> Prediksi: {r[1]} (p={r[2]:.2f})")
```

```python
# ===== 1) Mount Google Drive =====
from google.colab import drive
drive.mount('/content/drive')

# Path folder gambar di Google Drive
folder_path = "/content/drive/MyDrive/kumpulan_foto_angka"

# ===== 2) Import library =====
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageOps
import os

import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.utils import to_categorical

# ===== 3) Preprocess gambar agar mirip MNIST =====
def preprocess_to_mnist_28x28(img_pil):
    """
    Langkah:
    - Konversi ke grayscale
    - Auto-contrast
    - Invert (supaya digit jadi putih, background hitam)
    - Crop ke bounding box digit
    - Resize ke 20x20 (aspect ratio), lalu pad ke 28x28
    - Normalisasi ke [0,1] dan tambah axis channel
    """
    # Grayscale + autocontrast
    img = img_pil.convert('L')
    img = ImageOps.autocontrast(img)

    # Paksa invert (agar sesuai MNIST)
    img = ImageOps.invert(img)
    arr = np.array(img).astype(np.uint8)

    # Binarisasi ringan untuk cari bbox digit
    thr = np.mean(arr) * 0.8
    mask = arr > thr
    if mask.any():
        ys, xs = np.where(mask)
        y0, y1 = ys.min(), ys.max()
        x0, x1 = xs.min(), xs.max()
        img = img.crop((x0, y0, x1+1, y1+1))

    # Resize ke 20x20 dengan aspect ratio
    img.thumbnail((20, 20), Image.Resampling.LANCZOS)
    w, h = img.size

    # Pad ke 28x28 dan center
    canvas = Image.new('L', (28, 28), color=0)
    canvas.paste(img, ((28 - w)//2, (28 - h)//2))

    # Normalisasi ke [0,1] + tambah channel
    arr = np.array(canvas).astype('float32') / 255.0
    arr = arr[..., None]   # (28,28,1)

    return canvas, arr

# ===== 4) Load dataset MNIST =====
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalisasi & reshape
x_train = x_train.astype('float32') / 255.0
x_test  = x_test.astype('float32') / 255.0
x_train = np.expand_dims(x_train, -1)  # (60000,28,28,1)
x_test  = np.expand_dims(x_test, -1)

y_train = to_categorical(y_train, 10)
y_test  = to_categorical(y_test, 10)

# ===== 5) Definisi Model D3 (CNN) =====
model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    MaxPooling2D((2,2)),
    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D((2,2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(10, activation='softmax')
])

# Kompilasi
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# ===== 6) Training =====
history = model.fit(x_train, y_train,
                    epochs=5,
                    batch_size=128,
                    validation_split=0.1,
                    verbose=2)

# Evaluasi
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=0)
print(f"Akurasi di data uji MNIST: {test_acc:.4f}")

# ===== 7) Prediksi gambar dari folder Drive (grid 5x5) =====
file_list = [f for f in os.listdir(folder_path) if f.lower().endswith(('.jpg', '.png', '.jpeg'))]

rows, cols = 5, 5
fig, axes = plt.subplots(rows, cols, figsize=(12, 12))

for i, ax in enumerate(axes.flat):
    if i < len(file_list):
        fname = file_list[i]
        fpath = os.path.join(folder_path, fname)
        img_pil = Image.open(fpath)
        proc_img, arr = preprocess_to_mnist_28x28(img_pil)
        x_in = np.expand_dims(arr, axis=0)  # (1,28,28,1)

        pred = model.predict(x_in, verbose=0)
        digit = np.argmax(pred)

        ax.imshow(proc_img, cmap='gray')
        ax.set_title(f"{fname}\nPrediksi: {digit}", fontsize=9)
        ax.axis('off')
    else:
        ax.axis('off')  # jika file < 25, sisanya dikosongkan

plt.tight_layout()
plt.show()
```
